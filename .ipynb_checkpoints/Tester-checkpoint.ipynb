### **Data Preprocessing and Exploration**

The first step is loading in the dataset. I uploaded the file to Google Drive, and will now fetch the content and then load it into a Pandas `DataFrame`. This method does not require maintaining a local copy of the dataset.

```python
# Download URL for the dataset
url = "https://drive.google.com/uc?export=download&id=1yESNpmoY0qO5WOjfetgg7YV904_yPVen"

# Get the content from the URL
response = requests.get(url)
 
# Decode the content into a string
arff_content = response.content.decode("utf-8")

# Load the string content using arff.loadarff
data, meta = arff.loadarff(io.StringIO(arff_content))

# Convert to a Pandas DataFrame
df = pd.DataFrame(data)

# Inspect the first few rows
df.head()
```

I noticed the `Class` column is in bytes, so I must decode it to a string. Then I'll inspect the values of the column to confirm their names and ensure that there are only two classes present.

```python
# Decode from bytes to string
df["Class"] = df["Class"].str.decode('utf-8')

# Check Class values
df["Class"].value_counts()
```

The `Class` values must be converted to 0/1 for modeling purposes. I went with 0 for Cammeo and 1 for Osmancik.

```python
# Map values to 0/1
df["Class"] = df["Class"].map({"Cammeo": 0, "Osmancik": 1})

# Confirm Class values
df["Class"].value_counts()
```

I'll check for missing values in the data to see if any values need to be imputed and confirm the data type of the columns.

```python
# Print missing value counts and data types
print("Missing value counts:\n")
print(df.isna().sum())
print("\nData Types:\n")
print(df.dtypes)
```

Since there are no missing values and all the data types are correct, I'll proceed with data exploration. The summary below shows the general scale of the columns, which will determine the necessity of standardization.

```python
# Print summary of feature columns
df.drop(columns=["Class"]).describe()
```

The above data description reveals that the columns are indeed on different scales (e.g., `Perimeter` values range from around 300-400 and `Extent` values range from around 0-1). This means that the data should be standardized for optimal model performance. Before this step, though, I'll perform correlation analysis to get some preliminary insight into the relationship of the features with `Class` and assess potential multicollinearity.

```python
# Create correlation matrix
corr = df.corr()

# Create mask to hide redundant values (hide corrs between the same feature, only show each corr once)
mask = np.triu(np.ones_like(corr, dtype=bool))

# Create heatmap with Seaborn
sns.heatmap(corr, annot=True, cmap="coolwarm", mask=mask)

# Show the plot to visualize correlation matrix
plt.show()
```

I noticed that all features except `Extent` have a negative correlation with `Class`. This suggests that higher values of these features are more likely to be associated with class 0, which is Cammeo rice. The correlation heatmap also reveals that some features have a near-perfect correlation with one another (like `Area` and `Convex_Area`). This means that there is high redundancy among the features, and the dimensionality could be reduced by removing some features or by performing principal component analysis (PCA).

Another useful data exploration step before proceeding with modeling is to visualize the separation of features between classes. Well-separated features will allow simple classifiers linear discriminant analysis (LDA) and linear support vector machines (SVM) to form clear decision boundaries, while overlapping features require more complex nonlinear models or feature transformations.

```python
# Select feature columns (everything but "Class")
features = [col for col in df.columns if col != "Class"]

# Set grid size for plotting
rows, cols = 2, 4

# Set figure size
plt.figure(figsize=(18, 8))

# Loop through features
for i, feature in enumerate(features):
    plt.subplot(rows, cols, i + 1)   # Create plot for current feature
    sns.kdeplot(data=df, x=feature,  # Add density curves based on "Class"
        hue="Class", fill=True)
    plt.title(f"{feature} Distribution") 
    plt.xlabel(f"{feature} Measurement") 
    plt.ylabel("Density")

# Show the plots
plt.tight_layout()
plt.show()
```

The features have relatively good separation between the two classes, but there is some overlap that could require more complex boundaries than simple linear models.

To continue to prepare for modeling, I'll perform a train/test split on the data. It's important to perform this step before transformations such as standardization and dimensionality reduction to avoid data leakage, i.e., information from the training data biasing the testing data.

```python
# Set features and target
X = df[features]
y = df["Class"]

# Perform train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% test, 80% train
    random_state=0,     # Ensures reproducibility
    stratify=y          # Keeps class balance
)

# Insepct number of observation in each set
print(f"Observations in train set: {len(X_train)}")
print(f"Observations in test set: {len(X_test)}")
```

I'll then perform standardization on the data so each feaure is on the same scale.

```python
# Function to standardize train and test set: subtract mean, divide by standard deviation
def standardize_data(X_train, X_test):
    X_train_scaled = (X_train - X_train.mean()) / X_train.std()
    X_test_scaled = (X_test - X_test.mean()) / X_test.std()
    return X_train_scaled, X_test_scaled
    
X_train_scaled, X_test_scaled = standardize_data(X_train, X_test)
```

Next, I'll perform PCA on the standardized data to see if I can narrow down the feature list. I'll use the "maximizing spread" method we learning in class to perform PCA and then compute the variance explained by each component.

```python
# Function to perform PCA (assumes data is already centered/scaled)
def pca(X):
    # Compute covariance matrix
    n = X.shape[0]
    S = (1/n) * np.dot(X.T, X)

    # Perform eigen decomposition
    eigenvalues, eigenvectors = np.linalg.eig(S)

    # Sort eigenvalues (descending) and reorder eigenvectors
    idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]

    # Project data onto eigenvectors
    X_projected = np.dot(X, eigenvectors)

    return X_projected, eigenvalues, eigenvectors

# Perform PCA on the scaled training set
X_pca, eigenvalues, eigenvectors = pca(X_train_scaled)

# Compute the explained variance of each principal component (%, rounded to 3 decimals)
explained_variance_ratios = ((eigenvalues / np.sum(eigenvalues)) * 100).round(3)

# Compute cumulative variance after each principal component is considered
cumulative_variance = []
total = 0
for ratio in explained_variance_ratios:
    total += ratio
    cumulative_variance.append((total))

# Create DataFrame showing the results
df_variance = pd.DataFrame({
    "Explained Variance (%)": explained_variance_ratios,
    "Cumulative Variance (%)": cumulative_variance,
}, index=[f"PC{i}" for i in range(len(explained_variance_ratios))])

# Display the DataFrame
df_variance
```

The first 3 principal components capture 99.710% of the variance in the data, meaning there is a negligible difference between just using these 3 features compared to all 7. To explore these components further, I'll view the elements of each eigenvector. This will reveal how strongly each feature contributes to the principal component. Elements with a higher magnitude mean the feature contributes more strongly to the principal component's direction.

```python
# Make labels for the first 3 PCs
labels = [f"PC{i}" for i in range(3)]

# Create a DataFrame of the eigenvector elements
evec_elems = pd.DataFrame(eigenvectors[:, :3], index=features, columns=labels)

# Show absolute value of eigenvector elements rounded to 2 decimals
evec_elems.abs().round(3)
```

Analysis of each principal component:
- **PC0**: Dominated by `Area`, `Perimeter`, `Major_Axis_Length`, and `Convex_Area`, meaning it is related mostly to overall size features.
- **PC1**: Dominated by `Eccentricity`, `Minor_Axis_Length`, and `Extent`, meaning it is related to how elongated the rice grain is.
- **PC2**: Dominated by `Extent`, which is the ratio of how much the rice grain fills up the bounding box in the image.

I'll make some density plots again to observe the feature separation between the 3 principal components.

```python
# Create a DataFrame for the first 3 PCs
pc_scores_train = pd.DataFrame(X_pca[:, :3], columns=["PC0", "PC1", "PC2"])
pc_scores_train["Class"] = y_train.values

# Set grid size for plotting
rows, cols = 1, 3 

# Set figure size
plt.figure(figsize=(18, 5))

# Loop through principal components
for i, pc in enumerate(["PC0", "PC1", "PC2"]):
    plt.subplot(rows, cols, i + 1)           # Create plot for current PC
    sns.kdeplot(data=pc_scores_train, x=pc,  # Add density curves based on "Class"
        hue="Class", fill=True)               
    plt.title(f"{pc} Distribution by Class")
    plt.xlabel(f"{pc} Score")
    plt.ylabel("Density")

# Show the plots
plt.tight_layout()
plt.show()
```

The plots reveal that there is most class separation for PC0, indicating that this will be the most important feature for distinguishing between Cammeo and Osmancik rice.

### **Initial Model Testing**

Now that the data is set up, I'll run some initial models to see how they perform. I'll use built-in libraries for this initial testing phase, then implement the model myself for the one I wind up choosing. The classification models we learned in class are LDA, SVM, and neural networks, so these are the models I will test.

```python
# Define train set: data projected onto first 3 PCs
X_train_pcs = X_pca[:, :3]

# Define test set: Project using the same eigenvectors
X_test_pcs = np.dot(X_test_scaled, eigenvectors[:, :3])  

# Define LDA classifier
lda = LinearDiscriminantAnalysis()

# Fit the model on the training data
lda.fit(X_train_pcs, y_train)

# Predict on the test data
y_pred_lda = lda.predict(X_test_pcs)

# Print accuracy, classification report, and confusion matrix
print("LDA Accuracy:", accuracy_score(y_test, y_pred_lda))
print("LDA Classification Report:\n", classification_report(y_test, y_pred_lda))
print("LDA Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lda))
```

```python
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train_pcs, y_train)
y_pred_svm = svm.predict(X_test_pcs)

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("SVM Classification Report:\n", classification_report(y_test, y_pred_svm))
print("SVM Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score

# ----------------------------
# 1️⃣ Prepare data
# ----------------------------
# Use first 3 PCs
X_train_nn = X_pca[:, :1]
X_test_nn = np.dot(X_test_scaled, eigenvectors[:, :1])  

# ----------------------------
# 2️⃣ Build the model
# ----------------------------
model = Sequential([
    Dense(8, input_dim=1, activation='relu'),  # Hidden layer with 8 neurons
    Dense(4, activation='relu'),               # Optional second hidden layer
    Dense(1, activation='sigmoid')             # Output layer for binary classification
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# ----------------------------
# 3️⃣ Train the model
# ----------------------------
history = model.fit(X_train_nn, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=1)

# ----------------------------
# 4️⃣ Evaluate on test set
# ----------------------------
y_pred_prob = model.predict(X_test_nn)
y_pred = (y_pred_prob > 0.5).astype(int)  # Convert probabilities to 0/1

print("NN Accuracy:", accuracy_score(y_test, y_pred))
```
